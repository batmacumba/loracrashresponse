```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["10", "20", "40", "80", "160", "320"])
y = np.array([0.6666666666666666, 0.23333333333333334, 0.7666666666666667, 0.5666666666666667, 0.5, 0.8666666666666667])

plt.bar(x,y)
plt.xlabel('Number of Nodes')
plt.ylabel('PDR')
plt.title('PDR between Caller and Responder')
plt.savefig("pdr_caller_responder.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
